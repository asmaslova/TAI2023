# Условный оператор if

Сначала записывается часть if с условным выражением, далее могут следовать одна или более необязательных частей elif, и, наконец, необязательная часть else. Общая форма записи условной инструкции if выглядит следующим образом

if условие_1:
    действие_1
elif условие_2:
    действие_2
else:
    действие_else

```python
if not 1:
    print('true')
else:
    print('false')
```

```python
a = 3
if a < -5:
    print('Low')
elif -5 <= a <= 5:
    print('Mid')
else:
    print('High')
```

# Операторы continue и break

Оператор continue начинает следующий проход цикла, минуя оставшееся тело цикла (for или while)

```python
for i in range(10):
    if 4 <= i <= 6: 
        continue
        print(556554654)
    else:
        print(i)
```

Оператор break досрочно прерывает цикл

```python
for i in range(1, 10):
    if 3 <= i <= 7: 
        break
    else:
        print(i)
```

Извлечение среза. Оператор извлечения среза [a:b:h]
- a – начало среза
- b – окончание(не включительно)
- h - шаг среза
По умолчанию первый индекс равен 0, второй - длине строки, а шаг - 1.

```python
s = 'hello world'
print(s[0:3]) # первые 5 символов
print(s[6:])    # символы начиная с индекса 6 до конца 
print(s[::2])   # все символы через 1
print(s[::-1])  # все символы в обратном порядке
```

При вызове работе со строками необходимо помнить, что строки в Python относятся к категории неизменяемых объектов, то есть все функции и методы могут лишь создавать новую строку

- S.isdigit()	                Состоит ли строка из цифр
- S.isalpha()                 	Состоит ли строка из букв
- S.isalnum()               	Состоит ли строка из цифр или букв
- S.islower()                	Состоит ли строка из символов в нижнем регистре
